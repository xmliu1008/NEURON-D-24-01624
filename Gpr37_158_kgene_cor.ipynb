```python
import seaborn as sns
import numpy as np
import pandas as pd
import scanpy as sc
import matplotlib.pyplot as plt

sc.settings.verbosity = 3
sc.logging.print_header()
sc.settings.set_figure_params(dpi=80, facecolor='white')
```

```python
umap_bin= pd.read_csv('gaba-glut-umap_output_all.xls',sep='\t')
```

```python
from scipy.stats import pearsonr, spearmanr
```

```python
sns.set_theme(style="ticks")
sns.lmplot(data=umap_bin, x="Gpr37", y="Kcna1", scatter_kws=dict(s=25,edgecolor='none'))
plt.xlim((-1, -0.5))
plt.ylim((-1, -0.5))
spearman_coef, spearman_p = spearmanr(list(umap_bin['Gpr37']), list(umap_bin['Kcna1']))
print(f"Spearman correlation coefficient: {spearman_coef}, p-value: {spearman_p}")
```

```python
for i in K_channel_genes:
    slope,intercept = np.polyfit(list(umap_bin['Gpr37']), list(umap_bin[i]),1)
    spearman_coef, spearman_p = spearmanr(list(umap_bin['Gpr37']), list(umap_bin[i]))
    print(f"Gpr37_{i},r = {spearman_coef:4.2f}, p-value= {spearman_p:4.5f},slope = {slope:4.2f}, intercept= {intercept:4.2f}")
```
